```python
%load_ext autoreload
%autoreload 2
import numpy as np
import json
from cpmpy import * # pip3 install cpmpy
from cpmpy.solvers import CPM_ortools
```

```python
data = json.load(open('instances/type02/instance0.json'))
data
```

```python
def convert_sol(sol, newname):
    ret = dict()
    N = len(sol['list'])
    ret[newname] = []
    for i,e in enumerate(sol['list']):
        row = [0]*N
        row[e-1] = 1 # offset 1
        ret[newname].append(row)
    return ret
    
convert_sol(dict([('list', [3,1,4,2])]),'board')
```

```python
def convert_inst(data):
    binData = dict()
    binData['instance'] = data['instance']
    binData['size'] = data['size']
    N = data['size']
    binData['formatTemplate'] = dict()
    # N*N boolvars
    binData['formatTemplate']['board'] = [[dict([('high',1), ('low',0), ('type','dvar')])]*N ]*N
    binData['solutions'] = [convert_sol(sol,'board') for sol in data['solutions']]
    binData['nonSolutions'] = [convert_sol(sol,'board') for sol in data['nonSolutions']]
    binData['tests'] = [convert_sol(sol,'board') for sol in data['tests']]
    return binData

convert_inst(data)
```

```python
# convert all from 02 to 20
for i in range(10):
    indata = json.load(open(f"instances/type02/instance{i}.json"))
    outdata = convert_inst(indata)
    json.dump(outdata, fp=open(f"instances/type20/instance{i}.json", 'w'))
```
